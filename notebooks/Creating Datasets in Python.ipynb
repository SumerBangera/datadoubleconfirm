#### Creating hypothetical datasets: Over here I'm creating a dataset with five columns, namely id, gender of students, and scores of test 1, 2 and 3. There are a total of 50 male and 50 female students. Each test score follows its own normal distribution. 

```python
# importing necessary modules
import pandas as pd
import numpy as np
import random
```

```python
# setting a random seed
random.seed(1)
```

```python
gen = []
for i in range(0,50):
    gender = 'Male'
    gen.append((gender))
for i in range(51,101):
    gender = 'Female'
    gen.append((gender))
```

```python
# checking value of the first element
gen[0]
```

```python
# checking value of the 51st element
gen[50]
```

```python
# checking number of elements 
len(gen)
```

```python
data = []
for i in range(0,100):
    id = i+1
    gender = gen[i]
    test1 = round(np.random.normal(50, 10))
    test2 = round(np.random.normal(55, 8))
    test3 = round(np.random.normal(58, 9))
    data.append((id, gender, test1, test2, test3))
```

```python
data
```

```python
type(data)
```

```python
# looking at first row in list
data[0]
```

```python
# selecting first row fourth element
data[0][3]
```

```python
df = pd.DataFrame(np.array(data))
```

```python
df.columns = ['id','gender','test1','test2','test3']
```

```python
df
```

#### The dataset created is complete but usually this is not the case in real world. So I'm making some scores for test 1 and 2 to be missing.

```python
for i in range(0,9):
    df.loc[5*i,'test1'] = np.nan
    df.loc[6*i,'test2'] = np.nan
```
