#### This notebook demonstrates how one can make use of the [API function](https://www.tablebuilder.singstat.gov.sg/publicfacing/initApiList.action) to retrieve data tables from SingStat Table Builder. The example below is on Annual Population: M810381 - Residents By Age Group & Type Of Dwelling, Annual.  

```python
import json
import urllib.request
urlData = "https://www.tablebuilder.singstat.gov.sg/publicfacing/api/json/title/14910.json"
webURL = urllib.request.urlopen(urlData)
data = webURL.read()
print(data)
```

```python
df = json.loads(data)
```

```python
# note that there are two different kinds of data tables/ structures
df
```

```python
from pandas.io.json import json_normalize
```

```python
# population by age group for year 2000 to 2019
json_normalize(df['Level1'])
```

```python
# population by age group for year 2000 to 2019 by type of dwellings
json_normalize(df['Level2'])
```

```python
df2 = json_normalize(df['Level2'])
```

```python
# focusing only on the aggregate level for all age groups
df2[df2['level_1']=="Residents"]
```

```python
df2.to_csv('populationsg.csv')
```
